```python
import pandas as pd

# Ścieżki do plików (zakładam katalog ../data/ na podstawie poprzednich kroków)
path_scientists = '../data/scientists_with_identifiers.csv'
path_works = '../data/openalex_all_results_complete.csv'

try:
    # 1. Wczytanie plików
    print("⏳ Wczytywanie plików CSV...")
    df_scientists = pd.read_csv(path_scientists)
    df_works = pd.read_csv(path_works)

    print(f"✅ Wczytano AUTORÓW: {len(df_scientists)} rekordów")
    print(f"✅ Wczytano PRACE: {len(df_works)} rekordów")
    print("=" * 80)

    # 2. Analiza pliku z AUTORAMI (scientists_with_identifiers)
    print(f"📂 PLIK 1: SCIENTISTS_WITH_IDENTIFIERS (Pola: {len(df_scientists.columns)})")
    print(f"Lista pól: {df_scientists.columns.tolist()}")
    
    print("\nPrzykładowy rekord (Autorzy):")
    # Wyświetlamy transpozycję (.T) pierwszego wiersza dla lepszej czytelności
    display(df_scientists.head(1).T)
    
    print("-" * 80)

    # 3. Analiza pliku z PRACAMI (openalex_all_results_complete)
    print(f"📂 PLIK 2: OPENALEX_ALL_RESULTS_COMPLETE (Pola: {len(df_works.columns)})")
    print(f"Lista pól: {df_works.columns.tolist()}")
    
    print("\nPrzykładowy rekord (Prace):")
    display(df_works.head(1).T)
    
    print("=" * 80)
    
    # 4. Sprawdzenie kluczy do łączenia (ORCID)
    # To kluczowe, żeby zobaczyć czy format jest ten sam (np. czy jest 'https://orcid.org/' czy sam numer)
    print("🔍 SPRAWDZENIE KLUCZY (ORCID):")
    
    # Szukamy kolumny z 'orcid' w nazwie w obu plikach
    sci_orcid_col = [c for c in df_scientists.columns if 'orcid' in c.lower()][0]
    work_orcid_col = [c for c in df_works.columns if 'orcid' in c.lower()][0]
    
    print(f"Kolumna u autorów: '{sci_orcid_col}' -> Przykładowa wartość: {df_scientists[sci_orcid_col].iloc[0]}")
    print(f"Kolumna w pracach: '{work_orcid_col}' -> Przykładowa wartość: {df_works[work_orcid_col].iloc[0]}")

except FileNotFoundError as e:
    print(f"❌ Błąd: Nie znaleziono pliku. Sprawdź ścieżkę: {e}")
except IndexError:
    print("❌ Błąd: Nie znaleziono kolumny ORCID w którymś z plików.")
except Exception as e:
    print(f"❌ Wystąpił nieoczekiwany błąd: {e}")
```

```python
# Sprawdzamy statystyki brakujących abstraktów w załadowanym df_works
total_rows = len(df_works)

# Liczymy wiersze, gdzie abstract jest NaN (pusty)
missing_count = df_works['abstract'].isna().sum()
present_count = total_rows - missing_count
missing_percent = (missing_count / total_rows) * 100

print("📊 RAPORT DOSTĘPNOŚCI ABSTRAKTÓW")
print("=" * 40)
print(f"Całkowita liczba prac:       {total_rows}")
print(f"✅ Prace z abstraktem:       {present_count}")
print(f"❌ Prace BEZ abstraktu:      {missing_count}")
print("-" * 40)
print(f"📉 Procent braków:           {missing_percent:.2f}%")

# Wyświetlamy przykładowe tytuły prac, którym brakuje abstraktu (jeśli są takie)
if missing_count > 0:
    print("\nPrzykładowe 3 prace bez abstraktu (zostanie użyty sam tytuł):")
    no_abstract_sample = df_works[df_works['abstract'].isna()]['title'].head(3)
    for i, title in enumerate(no_abstract_sample, 1):
        print(f"{i}. {title}")
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# ================= KONFIGURACJA =================
TEST_LIMIT_WORKS = 3       # Ile prac bierzemy do testu
MAX_WORDS_LIMIT = 50       # Do ilu słów przycinamy tekst (ochrona przed wybuchem liczby n-gramów)
# Uwaga: Dla tekstu o długości N słów, liczba n-gramów to ok. N^2 / 2.
# Dla 50 słów -> ~1200 embeddingów na pracę.
# Dla 200 słów -> ~20,000 embeddingów na pracę.
# ================================================

# 1. Ładowanie modelu
# 'all-MiniLM-L6-v2' jest szybki i świetny do zadań semantycznych
model = SentenceTransformer('all-MiniLM-L6-v2')

# 2. Przygotowanie danych
subset = df_works.head(TEST_LIMIT_WORKS).copy()
all_ngrams_data = []

print(f"🔄 Przetwarzanie {len(subset)} prac...")

for idx, row in subset.iterrows():
    # A. Wybór źródła tekstu (Abstrakt > Tytuł)
    text = row.get('abstract')
    origin = 'Abstract'
    
    # Jeśli abstrakt jest pusty lub krótki, weź tytuł
    if pd.isna(text) or not isinstance(text, str) or len(text) < 10:
        text = row.get('title', '')
        origin = 'Title'
        
    if not text: continue # Pomijamy puste rekordy

    # B. Tokenizacja i limit słów
    words = text.split()
    if len(words) > MAX_WORDS_LIMIT:
        words = words[:MAX_WORDS_LIMIT]
    
    # Identyfikator do legendy wykresu
    # Skracamy tytuł, żeby legenda była czytelna
    full_title = row.get('title', row.get('display_name', f'Work {idx}'))
    short_label = (full_title[:30] + '..') if len(full_title) > 30 else full_title
    
    # C. Generowanie n-gramów (od 2 słów do całej długości)
    # n = długość okna (od 2 do liczby słów)
    for n in range(2, len(words) + 1):
        # i = indeks startowy
        for i in range(len(words) - n + 1):
            ngram_text = " ".join(words[i : i + n])
            
            all_ngrams_data.append({
                'work_id': idx,
                'work_label': short_label,
                'ngram_len': n,
                'text': ngram_text,
                'origin': origin
            })

# Tworzymy DataFrame z fragmentami
df_ngrams = pd.DataFrame(all_ngrams_data)
print(f"✅ Wygenerowano {len(df_ngrams)} fragmentów tekstu (n-gramów).")

# 3. Generowanie embeddingów
print("🚀 Generowanie embeddingów (może potrwać chwilę)...")
# encode zwraca numpy array
embeddings = model.encode(df_ngrams['text'].tolist(), show_progress_bar=True)

# 4. Redukcja wymiarów (PCA) do 2D
pca = PCA(n_components=2)
pca_result = pca.fit_transform(embeddings)

df_ngrams['x'] = pca_result[:, 0]
df_ngrams['y'] = pca_result[:, 1]

# 5. Wizualizacja
plt.figure(figsize=(14, 10))

# Rysujemy punkty, kolorując je wg pracy (work_label)
# style='origin' pokaże nam czy punkt pochodzi z tytułu czy abstraktu (choć w teście pewnie tylko jedno)
sns.scatterplot(
    data=df_ngrams,
    x='x', y='y',
    hue='work_label',
    palette='bright',
    alpha=0.6,
    s=30,
    edgecolor=None
)

# Dodanie podpisów dla kilku skrajnych punktów (opcjonalne, dla orientacji)
# Bierzemy np. najdłuższe n-gramy (czyli całe teksty), żeby zobaczyć gdzie leży "centrum" pracy
centers = df_ngrams.loc[df_ngrams.groupby('work_id')['ngram_len'].idxmax()]
for _, row in centers.iterrows():
    plt.text(row['x'], row['y'], f" FULL TEXT", fontsize=9, fontweight='bold')

plt.title(f'Przestrzeń semantyczna prac (PCA)\nKażdy punkt to fragment tekstu (od 2 słów do całości)', fontsize=14)
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(title='Praca', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()

# Wyświetlenie i zapisanie
plt.savefig('ngrams_pca_viz.png', dpi=300)
plt.show()

# 6. Zapisanie danych do CSV
output_file = 'ngrams_embeddings.csv'
# Zapisujemy bez wektorów (są duże), tylko metadane i współrzędne PCA
df_ngrams[['work_id', 'work_label', 'ngram_len', 'text', 'x', 'y']].to_csv(output_file, index=False)
print(f"💾 Zapisano dane wizualizacyjne do: {output_file}")
```
